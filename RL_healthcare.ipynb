```python
import gym
from gym import spaces
import numpy as np

class DiabetesEnv(gym.Env):
    def __init__(self, data):
        super(DiabetesEnv, self).__init__()

        self.data = data.reset_index(drop=True)
        self.current_step = 0

        # State: Glucose, HeartRate, Insulin, Time
        self.observation_space = spaces.Box(low=0, high=300, shape=(4,), dtype=np.float32)

        # Actions: 0 = do nothing, 1 = give insulin, 2 = suggest food
        self.action_space = spaces.Discrete(3)

    def reset(self):
        self.current_step = 0
        return self._get_state()

    def _get_state(self):
        row = self.data.iloc[self.current_step]
        return np.array([
            row['glucose_level'],
            row['heart_rate'],
            row['insulin_dose'],
            row['time']  # e.g. 0-23 for hours
        ], dtype=np.float32)

    def step(self, action):
        # Get current glucose to compute reward
        current_glucose = self.data.iloc[self.current_step]['glucose_level']

        # Reward logic
        reward = 0
        if action == 0:  # Do nothing
            reward = -1 if current_glucose > 180 or current_glucose < 70 else 1
        elif action == 1:  # Give insulin
            reward = 1 if current_glucose > 180 else -1
        elif action == 2:  # Suggest food
            reward = 1 if current_glucose < 70 else -1

        self.current_step += 1
        done = self.current_step >= len(self.data) - 1

        return self._get_state(), reward, done, {}
```

```python
import gymnasium as gym
from gymnasium import spaces
import numpy as np

class GlucoseEnv(gym.Env):  # ✅ Inherit from gym.Env (not plain object)
    def __init__(self, glucose_data):
        super(GlucoseEnv, self).__init__()

        self.glucose_data = glucose_data
        self.current_step = 0

        # Define action space (e.g., 0 = do nothing, 1 = insulin, etc.)
        self.action_space = spaces.Discrete(3)

        # Observation: just the glucose level (normalized)
        self.observation_space = spaces.Box(low=0, high=500, shape=(1,), dtype=np.float32)

    def reset(self, seed=None, options=None):
        self.current_step = 0
        glucose = self._get_glucose()
        return np.array([glucose], dtype=np.float32), {}

    def step(self, action):
        self.current_step += 1
        done = self.current_step >= len(self.glucose_data)

        glucose = self._get_glucose()

        # Reward function: penalize out-of-range values
        if glucose is None:
            reward = -1
        elif 80 <= glucose <= 140:
            reward = 1
        else:
            reward = -0.5

        obs = np.array([glucose], dtype=np.float32)
        info = {"glucose": glucose}

        return obs, reward, done, False, info

    def _get_glucose(self):
        while self.current_step < len(self.glucose_data):
            val = self.glucose_data[self.current_step]
            if not np.isnan(val):
                return val
            self.current_step += 1
        return 0.0
```

```python
import gym
from gym import spaces
import numpy as np

class GlucoseEnv(gym.Env):
    def __init__(self, glucose_data):
        super(GlucoseEnv, self).__init__()
        self.glucose_data = glucose_data
        self.current_step = 0

        # Observation = [glucose_now, trend]
        self.observation_space = spaces.Box(low=np.array([0, -500]), high=np.array([600, 500]), dtype=np.float32)

        # Actions: 0 = do nothing, 1 = administer insulin
        self.action_space = spaces.Discrete(2)

    def reset(self):
        self.current_step = 0
        glucose_now = self.glucose_data[self.current_step]
        glucose_prev = glucose_now
        trend = glucose_now - glucose_prev
        return np.array([glucose_now, trend], dtype=np.float32)

    def step(self, action):
        self.current_step += 1
        if self.current_step >= len(self.glucose_data):
            done = True
            return np.array([0, 0], dtype=np.float32), 0, done, {}

        glucose_now = self.glucose_data[self.current_step]
        glucose_prev = self.glucose_data[self.current_step - 1]
        trend = glucose_now - glucose_prev

        # Reward logic
        if 70 <= glucose_now <= 180:
            reward = 1
        elif glucose_now < 70 or glucose_now > 250:
            reward = -2
        else:
            reward = -0.5

        done = self.current_step >= len(self.glucose_data) - 1
        obs = np.array([glucose_now, trend], dtype=np.float32)

        return obs, reward, done, {}
```

```python
import gym
import numpy as np
from gym import spaces

class GlucoseEnv(gym.Env):
    def __init__(self, glucose_data, heart_rate_data, steps_data, carbs_data):
        super(GlucoseEnv, self).__init__()
        self.glucose_data = glucose_data
        self.heart_rate_data = heart_rate_data
        self.steps_data = steps_data
        self.carbs_data = carbs_data
        self.current_step = 0

        # Observation: glucose, trend, heart_rate, steps, carbs
        self.observation_space = spaces.Box(low=0, high=np.inf, shape=(5,), dtype=np.float32)
        self.action_space = spaces.Discrete(4)  # 0: none, 1: small, 2: medium, 3: large insulin

    def reset(self):
        self.current_step = 0
        return self._get_obs()

    def step(self, action):
        self.current_step += 1

        glucose_now = self.glucose_data[self.current_step]
        glucose_prev = self.glucose_data[self.current_step - 1] if self.current_step > 0 else glucose_now
        trend = glucose_now - glucose_prev
        heart_rate = self.heart_rate_data[self.current_step]
        steps = self.steps_data[self.current_step]
        carbs = self.carbs_data[self.current_step]

        # Reward shaping
        if 70 <= glucose_now <= 180:
            reward = 1
        elif glucose_now < 70 or glucose_now > 250:
            reward = -2
        else:
            reward = -0.5

        done = self.current_step >= len(self.glucose_data) - 1
        obs = np.array([glucose_now, trend, heart_rate, steps, carbs], dtype=np.float32)
        return obs, reward, done, {}

    def _get_obs(self):
        glucose_now = self.glucose_data[self.current_step]
        glucose_prev = self.glucose_data[self.current_step - 1] if self.current_step > 0 else glucose_now
        trend = glucose_now - glucose_prev
        heart_rate = self.heart_rate_data[self.current_step]
        steps = self.steps_data[self.current_step]
        carbs = self.carbs_data[self.current_step]
        return np.array([glucose_now, trend, heart_rate, steps, carbs], dtype=np.float32)
```
